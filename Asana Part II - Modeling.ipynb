# Part II:  Modeling

Using the final dataset created in part I, we want to determine which factors are strongly tied to user adoption.

```python
#import modules:
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
df = pd.read_csv("asana.csv")
```

```python
df.columns
```

### One-hot encoding

```python
dum = pd.get_dummies(df, columns=['creation_source', 'opted_in_to_mailing_list', 'enabled_for_marketing_drip', 'org_size', 'creation_month', 'creation_year'])
```

```python
dum.head(5)
```

```python
X = pd.concat([dum, df["last_session_creation_time"]], axis=1, join='inner')
```

```python
X.drop(["const", "Unnamed: 0", "user_id", "creation_time", "name", "email", "org_id", "invited_by_user_id", "email_domain", "adopted", "creation_time_num", "creation_last_login"], axis=1, inplace=True)
```

```python
X.head(5)
```

```python
y = df["adopted"]
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
mod = sm.OLS(y,X,missing='drop')
res = mod.fit()
```

```python
print res.summary()
```

The model output shows a warning that there may be strong multicollinearity, and almost all the coefficients are insignificant.  Let's double-check:

```python
df.corr()
```

We see that there is some high multicollinearity, specifically:
-  enabled for marketing drip & opted in email list
-  last session creation time & creation year
-  last session creation time & creation last login

Drop last session creation time and enabled for marketing drip (of the pairs, these had lower correlation with our target variable, adoption).

```python
dum2 = pd.get_dummies(df, columns=['creation_source', 'opted_in_to_mailing_list', 'org_size', 'creation_month'])
```

```python
X2 = dum2.drop(["Unnamed: 0", "user_id", "creation_time", "name", "email", "org_id", "invited_by_user_id", "email_domain", "adopted", "creation_time_num", 'enabled_for_marketing_drip', 'last_session_creation_time', 'creation_year'], axis=1)
```

```python
X2 = sm.add_constant(X2)
mod2 = sm.OLS(y,X2,missing='drop')
res2 = mod2.fit()
print res2.summary()
```

### Model Interpretation

-  Signups via guest invites and personal projects are both significant, which is not surprising because guest invites are most likely through work, and people working on personal projects may rely more heavily on a platform such as Asana to organize their work.  
-  Opting into the mailing list is also significant and positive, which makes sense because users are more likely to use Asana if they regularly receive emails regarding the platform, such as new updates, tips, etc.  
-  Smaller organization sizes (1 and 2) are correlated with adoption also.  This might be because smaller companies, especially startups, are more likely to adopt Asana across the company, whereas in bigger companies, there might just be isolated teams or individuals who use Asana.  
-  Some of the creation months are significant as well, with the summer months having negative coefficients.  This is a bit surprising as we wouldn't expect the month of account creation to impact adoption.

### Suggestions to the team based on analysis

Smaller organizations appear to have generally higher adoption rates, so perhaps more features and marketing campaigns could be catered toward smaller companies.  Opting into the mailing list is also quite important, so making an effort to get the user subscribed to it when they initially create their account will encourage adoption.  Signups via guest invites is significant, but this typically depends on whether an individual belongs to a company that already uses it.  Often, it is easier to persuade an individual to take action rather than an entire company to, so I would suggest first marketing to individuals who are working on personal projects.  I would also recommend the Product team to create some features that are catered more toward personal projects, since these types of users show higher adoption rates.
